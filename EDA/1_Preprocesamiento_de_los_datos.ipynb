```python
def unificar_data(ruta_data_eeg, ruta_data_emg, ruta_data_hand):
    
    data_eeg = pd.read_csv(ruta_data_eeg, sep=';', header=None)
    data_eeg[15] = data_eeg[14] # Machete mientas se soluciona el error
    
    data_emg = pd.read_csv(ruta_data_emg)
    
    data_guante = pd.read_csv(ruta_data_hand, sep=';')
    
    colombia_tz = pytz.timezone('America/Bogota')
    
    data_eeg['Fecha'] = pd.to_datetime(data_eeg[16], format='%d-%m-%Y %H:%M:%S %f')
    data_eeg['Fecha'] = data_eeg['Fecha'].apply(lambda x: colombia_tz.localize(x))
    data_eeg['Timestamp'] =data_eeg['Fecha'].apply(lambda x: x.timestamp())
    
    data_guante['Sample_Time'] = pd.to_datetime(data_guante['Sample_Time'])
    data_guante['Sample_Time'] = data_guante['Sample_Time'].apply(lambda x: colombia_tz.localize(x))
    data_guante['Timestamp'] =data_guante['Sample_Time'].apply(lambda x: x.timestamp())
    
    
    columns_data_unificada = [
                          'EEG_CHANNEL_1', 'EEG_CHANNEL_2', 'EEG_CHANNEL_3', 'EEG_CHANNEL_4', 'EEG_CHANNEL_5', 'EEG_CHANNEL_6', 'EEG_CHANNEL_7', 'EEG_CHANNEL_8', 
                          'EEG_CHANNEL_9', 'EEG_CHANNEL_10', 'EEG_CHANNEL_11', 'EEG_CHANNEL_12', 'EEG_CHANNEL_13', 'EEG_CHANNEL_14', 'EEG_CHANNEL_15', 'EEG_CHANNEL_16',
                          'EMG_CHANNEL_1', 'EMG_CHANNEL_2', 'EMG_CHANNEL_3', 'EMG_CHANNEL_4', 'EMG_CHANNEL_5', 'EMG_CHANNEL_6', 'EMG_CHANNEL_7', 'EMG_CHANNEL_8',
                          'THUMBFLEX1','THUMBFLEX2','INDEXFLEX1', 'INDEXFLEX2', 'INDEXFLEX3','MIDDLEFLEX1','MIDDLEFLEX2', 'MIDDLEFLEX3', 
                          'RINGFLEX1', 'RINGFLEX2', 'RINGFLEX3', 'LITTLEFLEX1', 'LITTLEFLEX2', 'LITTLEFLEX3', 'TIMESTAMP'                                                                    
                         ]

    data_unificada = pd.DataFrame(columns = columns_data_unificada)
    
    
    fs_eeg = 256

    EEG_CHANNEL_1 = np.array(data_eeg[0])
    EEG_CHANNEL_2 = np.array(data_eeg[1])
    EEG_CHANNEL_3 = np.array(data_eeg[2])
    EEG_CHANNEL_4 = np.array(data_eeg[3])
    EEG_CHANNEL_5 = np.array(data_eeg[4])
    EEG_CHANNEL_6 = np.array(data_eeg[5])
    EEG_CHANNEL_7 = np.array(data_eeg[6])
    EEG_CHANNEL_8 = np.array(data_eeg[7])
    EEG_CHANNEL_9 = np.array(data_eeg[8])
    EEG_CHANNEL_10 = np.array(data_eeg[9])
    EEG_CHANNEL_11 = np.array(data_eeg[10])
    EEG_CHANNEL_12 = np.array(data_eeg[11])
    EEG_CHANNEL_13 = np.array(data_eeg[12])
    EEG_CHANNEL_14 = np.array(data_eeg[13])
    EEG_CHANNEL_15 = np.array(data_eeg[14])
    EEG_CHANNEL_16 = np.array(data_eeg[15])

    time_eeg = np.arange(0, EEG_CHANNEL_1.shape[0]) / fs_eeg
    
    fs_emg = 200

    index_1_emg = np.argmin(abs(np.array(data_emg['Timestamp']) - data_eeg.iloc[0]['Timestamp']))
    print(f'Index 1 eeg-emg: {index_1_emg}')
    desfase = str(abs(np.array(data_emg['Timestamp']) - data_eeg.iloc[0]['Timestamp'])[index_1_emg])
    print(f'Desfase: {desfase}')

    index_2_emg = np.argmin(abs(np.array(data_emg['Timestamp']) - data_eeg.iloc[-1]['Timestamp']))
    print(f'Index 2 eeg-emg: {index_2_emg}')
    desfase = str(abs(np.array(data_emg['Timestamp']) - data_eeg.iloc[-1]['Timestamp'])[index_2_emg])
    print(f'Desfase: {desfase}')

    EMG_CHANNEL_1 = np.array(data_emg['Channel 0'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_2 = np.array(data_emg['Channel 1'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_3 = np.array(data_emg['Channel 2'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_4 = np.array(data_emg['Channel 3'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_5 = np.array(data_emg['Channel 4'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_6 = np.array(data_emg['Channel 5'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_7 = np.array(data_emg['Channel 6'])[index_1_emg:index_2_emg+1]
    EMG_CHANNEL_8 = np.array(data_emg['Channel 7'])[index_1_emg:index_2_emg+1]

    time_emg = np.arange(0, EMG_CHANNEL_1.shape[0]) / fs_emg

    EMG_CHANNEL_1_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_1)
    EMG_CHANNEL_2_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_2)
    EMG_CHANNEL_3_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_3)
    EMG_CHANNEL_4_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_4)
    EMG_CHANNEL_5_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_5)
    EMG_CHANNEL_6_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_6)
    EMG_CHANNEL_7_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_7)
    EMG_CHANNEL_8_OVERSAMPLED = np.interp(time_eeg, time_emg, EMG_CHANNEL_8)
    
    
    fs_hand = 2.3

    index_1_hand = np.argmin(abs(np.array(data_guante['Timestamp']) - data_eeg.iloc[0]['Timestamp']))
    print(f'Index 1 eeg-hand: {index_1_hand}')
    desfase = str(abs(np.array(data_guante['Timestamp']) - data_eeg.iloc[0]['Timestamp'])[index_1_hand])
    print(f'Desfase: {desfase}')

    index_2_hand = np.argmin(abs(np.array(data_guante['Timestamp']) - data_eeg.iloc[-1]['Timestamp']))
    print(f'Index 2 eeg-hand: {index_2_hand}')
    desfase = str(abs(np.array(data_guante['Timestamp']) - data_eeg.iloc[-1]['Timestamp'])[index_2_hand])
    print(f'Desfase: {desfase}')

    data_guante['THUMBFLEX1']= data_guante['THUMBFLEX1'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['THUMBFLEX2']= data_guante['THUMBFLEX2'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['INDEXFLEX1']= data_guante['INDEXFLEX1'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['INDEXFLEX2']= data_guante['INDEXFLEX2'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['INDEXFLEX3']= data_guante['INDEXFLEX3'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['MIDDLEFLEX1']= data_guante['MIDDLEFLEX1'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['MIDDLEFLEX2']= data_guante['MIDDLEFLEX2'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['MIDDLEFLEX3']= data_guante['MIDDLEFLEX3'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['RINGFLEX1']= data_guante['RINGFLEX1'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['RINGFLEX2']= data_guante['RINGFLEX2'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['RINGFLEX3']= data_guante['RINGFLEX3'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['LITTLEFLEX1']= data_guante['LITTLEFLEX1'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['LITTLEFLEX2']= data_guante['LITTLEFLEX2'].apply(lambda x: float(x.replace(',', '.')))
    data_guante['LITTLEFLEX3']= data_guante['LITTLEFLEX3'].apply(lambda x: float(x.replace(',', '.')))

    THUMBFLEX1 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    THUMBFLEX2 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    INDEXFLEX1 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    INDEXFLEX2 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    INDEXFLEX3 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    MIDDLEFLEX1 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    MIDDLEFLEX2 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    MIDDLEFLEX3  = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    RINGFLEX1 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    RINGFLEX2 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    RINGFLEX3 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    LITTLEFLEX1 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    LITTLEFLEX2 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]
    LITTLEFLEX3 = np.array(data_guante['THUMBFLEX1'])[index_1_hand:index_2_hand+1]

    time_hand = np.arange(0, THUMBFLEX1.shape[0]) / fs_hand

    THUMBFLEX1_OVERSAMPLED = np.interp(time_eeg, time_hand, THUMBFLEX1)
    THUMBFLEX2_OVERSAMPLED = np.interp(time_eeg, time_hand, THUMBFLEX2)
    INDEXFLEX1_OVERSAMPLED = np.interp(time_eeg, time_hand, INDEXFLEX1)
    INDEXFLEX2_OVERSAMPLED = np.interp(time_eeg, time_hand, INDEXFLEX2)
    INDEXFLEX3_OVERSAMPLED = np.interp(time_eeg, time_hand, INDEXFLEX3)
    MIDDLEFLEX1_OVERSAMPLED = np.interp(time_eeg, time_hand, MIDDLEFLEX1)
    MIDDLEFLEX2_OVERSAMPLED = np.interp(time_eeg, time_hand, MIDDLEFLEX2)
    MIDDLEFLEX3_OVERSAMPLED = np.interp(time_eeg, time_hand, MIDDLEFLEX3)
    RINGFLEX1_OVERSAMPLED = np.interp(time_eeg, time_hand, RINGFLEX1)
    RINGFLEX2_OVERSAMPLED = np.interp(time_eeg, time_hand, RINGFLEX2)
    RINGFLEX3_OVERSAMPLED = np.interp(time_eeg, time_hand, RINGFLEX3)
    LITTLEFLEX1_OVERSAMPLED = np.interp(time_eeg, time_hand, LITTLEFLEX1)
    LITTLEFLEX2_OVERSAMPLED = np.interp(time_eeg, time_hand, LITTLEFLEX2)
    LITTLEFLEX3_OVERSAMPLED = np.interp(time_eeg, time_hand, LITTLEFLEX3)

    
    data_unificada['EEG_CHANNEL_1'] = EEG_CHANNEL_1
    data_unificada['EEG_CHANNEL_2'] = EEG_CHANNEL_2
    data_unificada['EEG_CHANNEL_3'] = EEG_CHANNEL_3
    data_unificada['EEG_CHANNEL_4'] = EEG_CHANNEL_4
    data_unificada['EEG_CHANNEL_5'] = EEG_CHANNEL_5
    data_unificada['EEG_CHANNEL_6'] = EEG_CHANNEL_6
    data_unificada['EEG_CHANNEL_7'] = EEG_CHANNEL_7
    data_unificada['EEG_CHANNEL_8'] = EEG_CHANNEL_8
    data_unificada['EEG_CHANNEL_9'] = EEG_CHANNEL_9
    data_unificada['EEG_CHANNEL_10'] = EEG_CHANNEL_10
    data_unificada['EEG_CHANNEL_11'] = EEG_CHANNEL_11
    data_unificada['EEG_CHANNEL_12'] = EEG_CHANNEL_12
    data_unificada['EEG_CHANNEL_13'] = EEG_CHANNEL_13
    data_unificada['EEG_CHANNEL_14'] = EEG_CHANNEL_14
    data_unificada['EEG_CHANNEL_15'] = EEG_CHANNEL_15
    data_unificada['EEG_CHANNEL_16'] = EEG_CHANNEL_16
    data_unificada['EMG_CHANNEL_1'] = EMG_CHANNEL_1_OVERSAMPLED
    data_unificada['EMG_CHANNEL_2'] = EMG_CHANNEL_2_OVERSAMPLED
    data_unificada['EMG_CHANNEL_3'] = EMG_CHANNEL_3_OVERSAMPLED
    data_unificada['EMG_CHANNEL_4'] = EMG_CHANNEL_4_OVERSAMPLED
    data_unificada['EMG_CHANNEL_5'] = EMG_CHANNEL_5_OVERSAMPLED
    data_unificada['EMG_CHANNEL_6'] = EMG_CHANNEL_6_OVERSAMPLED
    data_unificada['EMG_CHANNEL_7'] = EMG_CHANNEL_7_OVERSAMPLED
    data_unificada['EMG_CHANNEL_8'] = EMG_CHANNEL_8_OVERSAMPLED
    data_unificada['THUMBFLEX1'] = THUMBFLEX1_OVERSAMPLED
    data_unificada['THUMBFLEX2'] = THUMBFLEX2_OVERSAMPLED
    data_unificada['INDEXFLEX1'] = INDEXFLEX1_OVERSAMPLED
    data_unificada['INDEXFLEX2'] = INDEXFLEX2_OVERSAMPLED
    data_unificada['INDEXFLEX3'] = INDEXFLEX3_OVERSAMPLED
    data_unificada['MIDDLEFLEX1'] = MIDDLEFLEX1_OVERSAMPLED
    data_unificada['MIDDLEFLEX2'] = MIDDLEFLEX2_OVERSAMPLED
    data_unificada['MIDDLEFLEX3'] = MIDDLEFLEX3_OVERSAMPLED
    data_unificada['RINGFLEX1'] = RINGFLEX1_OVERSAMPLED
    data_unificada['RINGFLEX2'] = RINGFLEX2_OVERSAMPLED
    data_unificada['RINGFLEX3'] = RINGFLEX3_OVERSAMPLED
    data_unificada['LITTLEFLEX1'] = LITTLEFLEX1_OVERSAMPLED
    data_unificada['LITTLEFLEX2'] = LITTLEFLEX2_OVERSAMPLED
    data_unificada['LITTLEFLEX3'] = LITTLEFLEX3_OVERSAMPLED
    data_unificada['TIMESTAMP'] = time_eeg
    
    return data_unificada
    
```
